```python
df_1 = df.dropna()
```

```python
df_1.shape
```

### THÊM ĐẶC TRƯNG MỚI

```python
df_1['Age'] = 2021 - df_1.loc[:,'Year_Birth']
```

```python
df_1["Dt_Customer"] = pd.to_datetime(df_1["Dt_Customer"])
dates = []
for i in df_1["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
#Dates of the newest and oldest recorded customer
print("The newest customer's enrolment date in therecords:",max(dates))
print("The oldest customer's enrolment date in the records:",min(dates))
```

```python
days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
df_1["Customer_For"] = days
```

```python
df_1.head()
```

```python
day = []
for i in df_1['Customer_For']:
    day.append(int(str(i).split()[0]))
```

```python
df_1["Customer_For"] = day
```

```python
df_1["Spent"] = df_1["MntWines"]+ df_1["MntFruits"]+ df_1["MntMeatProducts"]+ df_1["MntFishProducts"]+ df_1["MntSweetProducts"]+ df_1["MntGoldProds"]
```

```python
df_1.head()
```

### MÃ HÓA DỮ LIỆU

```python
df_2 = df_1.copy()
```

```python
df_2.Marital_Status.value_counts()
```

```python
dic = {'Single':'Alone','Together':'Married','Married':'Married','Divorced':'Alone', 'Widow':'Alone', 'Alone':'Alone',
      'Absurd':'Alone', 'YOLO':'Alone'}
```

```python
df_2['Marital_Status'] = df_1['Marital_Status'].map(dic)
```

```python
df_2.Education.value_counts()
```

```python
df_2["Education"]=df_1["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

```python
df_2.Education.value_counts()
```

```python
df_2['Marital_Status'] = df_2['Marital_Status'].map({'Alone':0,"Married":1})
```

```python
df_2['Education'] = df_2['Education'].map({'Undergraduate':0,"Postgraduate":1,'Graduate':2})
```

```python
df_2.head()
```

```python
variable_drop = ['ID','Dt_Customer',"Z_CostContact", "Z_Revenue",'Year_Birth']
df_2 = df_2.drop(variable_drop,axis = 1)
```

```python
df_2.tail()
```

```python
df_2.columns
```

```python
variable_drop = ['AcceptedCmp3',
       'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2']
df_2 = df_2.drop(variable_drop,axis = 1)
```

```python
df_2.shape
```

### CHUẨN HÓA DỮ LIỆU

```python
df_3 = df_2.copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
list_col = list(df_3.columns)
list_col =  [col for col in list_col if col != 'Education' and col != 'Marital_Status']
```

```python
scale = StandardScaler().fit(df_3[list_col])
```

```python
df_3[list_col] = scale.transform(df_3[list_col])
```

```python
df_3.describe()
```

### GIẢM CHIỀU DỮ LIỆU

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components=3).fit(df_3)
```

```python
PCA_ds = pd.DataFrame(pca.transform(df_3), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

## HUẤN LUYỆN MÔ HÌNH - ĐÁNH GIÁ

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
elbrow = KElbowVisualizer(KMeans(),k = 10).fit(PCA_ds)
elbrow.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
from matplotlib.colors import ListedColormap
```

```python
AC = AgglomerativeClustering(n_clusters=4)
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC
df_3["Clusters"]= yhat_AC
```

```python
from matplotlib import colors
cmap = colors.ListedColormap(["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"])
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o',cmap = cmap)
ax.set_title("The Plot Of The Clusters")
plt.show()
```

## ĐÁNH GIÁ MÔ HÌNH

```python
#Plotting countplot of clusters
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=df_3["Clusters"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```
