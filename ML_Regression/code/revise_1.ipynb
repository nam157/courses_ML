```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
def rmse(model,X,y):
    cv = KFold(n_splits=10,shuffle=True,random_state=0).get_n_splits(X.values)
    score = cross_val_score(model,X,y,cv = cv,scoring='neg_mean_squared_error')
    score = np.sqrt(-score)
    return score.mean(),score.std()
```

```python
def get_array(X,y):
    one = np.ones((X.shape[0],1))
    X = np.array(X)
    return np.concatenate((one,X),axis = 1),np.array(y)
```

```python
def predict(X,w):
    return X.dot(w)
```

```python
def loss(X,y,w):
    m,n = X.shape
    y_pr = predict(X,w)
    error = y_pr - y
    return np.sum(error**2)
```

```python
def derivative(X,y,w):
    y_pr = predict(X,w)
    error = y_pr - y
    return -2 * error.dot(X) 
```

```python
def update_weight(X,y,w,learing_rate,iters):
    costs = []
    for _ in range(iters):
        for i in range(len(w)):
            de = derivative(X[:,i],y,w[i])
            w[i] = w[i] - learing_rate*de
        cost = loss(X,y,w)
        costs.append(cost)
    return w,costs
```

```python
X,y= get_array(X_train_lasso,y_train)
weight = np.zeros(X.shape[1])
learing_rate = 0.000001
iters = 1000
weights,cost = update_weight(X,y,weight,learing_rate,iters)
```
